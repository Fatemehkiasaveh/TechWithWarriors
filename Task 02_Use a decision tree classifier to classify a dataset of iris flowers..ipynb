Task 02

Use a decision tree classifier to classify a dataset
of iris flowers.

Task Details: Implement a decision tree classifier to
categorize iris flowers into different species based on
their features.

Dataset: Iris Dataset

```python
import seaborn as sns

iris = sns.load_dataset("iris")
iris
```

```python
iris.species.value_counts()
```

```python
iris.isna().sum()
```

```python
x = iris.drop(columns='species')
y = iris['species']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=44)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()

dt.fit(x_train, y_train)
```

```python
dt.score(x_train, y_train)
```

```python
dt.score(x_test, y_test)
```

```python
dt1 = DecisionTreeClassifier(max_depth=8)
dt1.fit(x_train, y_train)
```

```python
dt1.score(x_train, y_train)
```

```python
dt1.score(x_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

dt2 = DecisionTreeClassifier()
parameters = {'max_depth':[3,4,5,6,7], 'max_leaf_nodes':[2,3,4,5,6]}
dt2 = GridSearchCV(dt2, parameters)
dt2.fit(x_train, y_train)
```

```python
dt2.score(x_train, y_train)
```

```python
dt2.score(x_test, y_test)
```

```python
dt2.best_params_
```

```python
dt3 = DecisionTreeClassifier(max_depth=3, max_leaf_nodes=3)
dt3.fit(x_train, y_train)
```

```python
dt3.score(x_train, y_train)
```

```python
dt3.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = dt3.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import numpy as np
```

```python
# Calculate per-class metrics
precision = np.diag(cm) / np.sum(cm, axis=0)
recall = np.diag(cm) / np.sum(cm, axis=1)
f1_score = 2 * (precision * recall) / (precision + recall)

print("Precision per class:", precision)
print("Recall per class:", recall)
print("F1 score per class:", f1_score)
```

```python
y_pred_total = dt3.predict(x)

cm_total = confusion_matrix(y, y_pred_total)

precision = np.diag(cm_total) / np.sum(cm_total, axis=0)
recall = np.diag(cm_total) / np.sum(cm_total, axis=1)
f1_score = 2 * (precision * recall) / (precision + recall)

print("Precision per class:", precision)
print("Recall per class:", recall)
print("F1 score per class:", f1_score)
```
